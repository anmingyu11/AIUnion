```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
stdScaler = StandardScaler()
```

```python
X_iris_std = stdScaler.fit_transform(X_iris)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(4):
    plt.plot(X_iris_std[:,i])
plt.show()
```

```python
def plot_scaled_data(X_data):
    maxV = X_data[0].max()
    minV = X_data[0].min()
    
    for i in range(4):
        maxV = max(maxV,max(X_data[i]))
        minV = min(minV,min(X_data[i]))
    print('max:',maxV)
    print('min:',minV)
    for i in range(4):
        plt.plot(X_data[:,i])
    plt.show()
```

```python
plot_scaled_data(X_iris_std)
```
