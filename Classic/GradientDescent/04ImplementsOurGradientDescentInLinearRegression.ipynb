## 使用梯度下降法训练

```python
def J(theta, X_b, y):
    try:
        return np.sum((y - X_b.dot(theta))**2) / len(X_b)
    except:
        return float('inf')
```

```python
def dJ(theta,X_b,y): #这个函数我写出了无穷的错误
    m = len(theta)
    res = np.empty(m)
    res[0] = np.sum(X_b.dot(theta) - y)
    for i in range(1, m):
        res[i] = (X_b.dot(theta) -y) .dot( X_b[:,i])
    print('len theta : ' , len(theta))
    print('len X_b : ' , len(X_b))
    return res * 2 / len(X_b)
```

```python
def gradient_descent(X_b,y,init_theta,eta,n_iters=1e4,epsilon=1e-8):
    theta = init_theta
    cur_iter=0
    
    while cur_iter < n_iters:
        gradient = dJ(theta,X_b,y)
        last_theta = theta
        theta = theta - eta * gradient
        
        if(abs(J(theta,X_b,y) - J(last_theta,X_b,y)) < epsilon):
            break
        
        cur_iter += 1
    
    return theta
```

```python
X_b = np.hstack([np.ones((len(x),1)),x.reshape(-1,1)]) # 生成一个n行1列的 Θ0列向量,再将x转化为n*1的列向量
initial_theta = np.zeros(X_b.shape[1])
```
